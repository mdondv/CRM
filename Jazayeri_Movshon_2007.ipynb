```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import norm, gamma
```

Encoding stage: the sensory representation has a Gaussian probability density function centered at the true direction of motion

Decoding stage: computess sensory likelihood by multiplying the sensory representation with a weighting profile

```python
theta_stim = []
theta = 90
while(theta >= -90): #np.arange
    theta_stim.append(theta)
    theta -= 1
    
dict = {'θ': 3*theta_stim}
df = pd.DataFrame(dict)

df['coh'] = 0

for i in range(3*len(theta_stim)):
    if(i < len(theta_stim)):
        df.loc[i,['coh']] = 0.03
    elif(len(theta_stim)-1 < i < 2*len(theta_stim)):
        df.loc[i,['coh']] = 0.06
    else:
        df.loc[i,['coh']] = 0.12
        
df['sens_rep'] = 0 #sensory representation has a gaussian probability centered at the true direction of motion

for i in range(len(df)):
    df.loc[i,['sens_rep']] = np.random.normal(df.loc[i,['θ']], df.loc[i,['coh']])[0]

df['prop_CW'] = 0
k = .75

for i in range(len(df)):
    df.loc[i,['prop_CW']] = 1 - norm.cdf(0, df.loc[i,['θ']], k*1 / df.loc[i,['coh']])
    
df['θ_est'] = 0
x_values = np.linspace(-40, 40, 300)
a = 4

for i in range(len(df)):
        y = norm.pdf(x_values, df.loc[i,['θ']], k*1 / df.loc[i,['coh']]) * gamma.pdf(x_values, a, scale=6)
        df.loc[i,['θ_est']] = x_values[y == max(y)]

df['choice'] = 0
    
for i in range(len(df)):
    df.loc[i, 'choice'] = np.random.choice(['CW', 'CCW'], p=[df.loc[i, 'prop_CW'], 1 - df.loc[i, 'prop_CW']])

df['choice_bin'] = 0 

conditions = [
    (df['choice'] == 'CW'),
    (df['choice'] == 'CCW')
    ]

values = [1, 0]

# create a new column and use np.select to assign values to it using our lists as arguments
df['choice_bin'] = np.select(conditions, values)

df
```

```python
plt.xlim(-25, 25)

plt.plot(df.loc[0:len(theta_stim)-1, ['θ']], df.loc[0:len(theta_stim)-1, ['prop_CW']])
plt.plot(df.loc[len(theta_stim): 2*(len(theta_stim)-1), ['θ']], df.loc[len(theta_stim): 2*(len(theta_stim)-1), ['prop_CW']])
plt.plot(df.loc[2*(len(theta_stim)): 3*(len(theta_stim)-1), ['θ']], df.loc[2*(len(theta_stim)): 3*(len(theta_stim)-1), ['prop_CW']])

plt.axvline(x=0, color='k', linestyle='--', linewidth=1)
plt.axhline(y=0.5, color='k', linestyle='--', linewidth=1)

plt.ylabel("Proportion judged CW")
plt.xlabel("Direction (degrees)")
plt.show()
```
